```python
import pandas as pd
import struct
```

```python
file_path = "D:\SierraChart\Data\MarketDepthData\ESM24-CME.2024-04-08.depth"


def read_scid_file(file_path):
    record_format = '8i'  # Format: 8 integerów
    record_size = struct.calcsize(record_format)
    
    data = []

    with open(file_path, 'rb') as file:
        while True:
            record = file.read(record_size)
            if len(record) < record_size:
                break
            unpacked_record = struct.unpack(record_format, record)
            data.append(unpacked_record)
    
    # Tworzenie DataFrame z danych
    df = pd.DataFrame(data, columns=['Field1', 'Field2', 'Field3', 'Field4', 'Field5', 'Field6', 'Field7', 'Field8'])
    return df

# Przykładowe użycie
df = read_scid_file(file_path)

# Wyświetlenie pierwszych kilku wierszy DataFrame
df.head()
```

```python
# Próbujemy odczytać pierwszy rekord jako różne typy danych
record = data[0]

# Przykładowe struktury do odczytu
try:
    unpacked_record = struct.unpack('8i', record)  # 8 integerów
    print("8 integers:", unpacked_record)
except:
    pass

try:
    unpacked_record = struct.unpack('4f', record)  # 4 floaty
    print("4 floats:", unpacked_record)
except:
    pass

try:
    unpacked_record = struct.unpack('2d', record)  # 2 double
    print("2 doubles:", unpacked_record)
except:
    pass
```
